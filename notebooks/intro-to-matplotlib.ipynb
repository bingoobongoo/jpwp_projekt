```python
x = [1, 2 , 3 , 4]
y = [11, 22, 33, 44]
plt.plot(x,y)
```

```python
# 1st method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x,y)
```

## Matplotlib workflow

```python
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(8,8)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple Plot',
       xlabel='x-axis',
       ylabel='y-axis')

# 5. Save and show
fig.savefig('simple_plot.png')
```

```python
nut_butter_prices = {'Almond butter':10,
                     'Peanut butter':8 ,
                     'Cashew butter':12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),
        nut_butter_prices.values())
ax.set(title='Butter Prices',
       ylabel='Price ($)')
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
heart_disease_df = pd.read_csv('/content/heart-disease.csv')
heart_disease_df
```

```python
heart_disease_df['age'].plot(kind='hist', bins=50)
```

```python
heart_disease_df.plot.hist(figsize=(10,30), subplots=True)
```

```python
over_50 = heart_disease_df[heart_disease_df['age'] > 50]
over_50
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol');
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol Level')

ax.legend(*scatter.legend_elements(), title='Target')

ax.axhline(y=over_50['chol'].mean(),
           linestyle='--')
```

```python
fig, ax = plt.subplots(figsize=(10,14),
                       nrows=2,
                       ncols=1)
ax[0].scatter(x=over_50['age'],
              y=over_50['thalach'],
              c=over_50['target'])

ax[0].set(title='Heart Disease Thalach',
          xlabel='Age',
          ylabel='Thalach',
          ylim=[50, 250])

ax[0].axhline(y=over_50['thalach'].mean(),
              linestyle='--',
              color='red')

ax[0].legend(*scatter.legend_elements(), title='Target')

ax[1].scatter(x=over_50['age'],
              y=over_50['chol'],
              c=over_50['target'])

ax[1].set(title='Heart Disease and Cholesterol Levels',
          xlabel='Age',
          ylabel='Cholesterol Level')

ax[1].legend(*scatter.legend_elements(), title='Target')

ax[1].axhline(y=over_50['chol'].mean(),
              linestyle='--')

fig.suptitle('HealthData', fontsize=16, fontweight='bold')
```
