```python
import pandas as pd
import numpy as np
```

```python
pokemon_df = pd.read_csv('../data/pokemon.csv')
```

```python
pokemon_df
```

```python
boolean_names = []
for name in pokemon_df['Name']:
    if name.find(' ') == -1:
        boolean_names.append(False)
    else:
        boolean_names.append(True)
```

```python
dropped = pokemon_df.drop(
    pokemon_df.index[boolean_names])
```

```python
dropped = pokemon_df.drop(labels=['#'], axis=1)
```

```python
dropped
```

```python
len(dropped)
```

```python
dropped.to_csv('../data/pokemon.csv')
```
